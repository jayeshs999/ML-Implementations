```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
train_set= datasets.load_boston()
```

```python
train_set
```

```python
X,y=train_set.data,train_set.target.reshape(len(train_set.target),1)
X=np.hstack((X,np.power(X[:,[1,2,7,8,9,10,11,12]],0.5),np.power(X[:,[1,2,7,8,9,10,11,12]],2),np.power(X[:,[1,2,7,8,9,10,11,12]],3)))
X=np.hstack((np.ones((len(X),1)),X))
X.shape
```

```python
print(np.linalg.det(X.T.dot(X)))
```

```python
y.shape
```

```python
theta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
theta
```

```python
(X.dot(theta)-y)
```

```python
np.sum(np.power((X.dot(theta)-y),2)/1012)
```

```python
plt.scatter(np.arange(0,len(X.dot(theta))),X.dot(theta)-y)
```
