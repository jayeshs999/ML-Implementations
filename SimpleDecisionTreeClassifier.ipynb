```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

```python
X,y=load_breast_cancer(return_X_y=True)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=9,test_size=0.3)
```

```python
model=DecisionTreeClassifier(random_state=43)
```

```python
model=model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.feature_importances_
```

```python
plt.figure(figsize=(20,10))
plt.barh(load_breast_cancer().feature_names,model.feature_importances_,height=0.8)
```
